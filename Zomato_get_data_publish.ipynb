```python
#creating the url for extraction of the data using the API
accept = 'application/json'
values = {'entity_id': 302 ,
          'entity_type': 'city' ,
          'sort' : 'rating',
          'order' : 'desc' ,
          'start' : 0 ,
          'count' : 20 }
headers = {'user-key': user_key,
           'accept':accept,
           'User-agent' : 'curl/7.43.0' }

data = urllib.parse.urlencode(values)
print(str(data))
```

```python
#List of cities to be used for extrcation, the city_id is used using the Zomato API
listCities2 = [39,40,11387,11409,11338,24,8,27]
```

```python
# define the dataframe columns
column_names = ['restaurant_name', 'restaurant_rating', 'resturant_city','latitude','longitude','cost4two','cuisines','zipcode'] 

# instantiate the dataframe
df1 = pd.DataFrame(columns=column_names)

df=pd.DataFrame
for each_city in listCities2:
    url = 'https://developers.zomato.com/api/v2.1/search'
    values['entity_id'] = each_city
    counter = 0
    listToSave = []
    strCity = ""
    while (counter < 100):
        values['start'] = counter
        data = urllib.parse.urlencode(values)
        print("\n" + str(counter) + "\n")
        url = url + '?' + data
        req = urllib.request.Request(url, None, headers)
        response = urllib.request.urlopen(req)
        the_page = response.read()
        dict_response = json.loads(the_page)
        if 'restaurants' in dict_response:
            listRestaurants = dict_response['restaurants']
            for each_restaurant in listRestaurants:
                name=each_restaurant['restaurant']['name']
                res_rating = each_restaurant['restaurant']['user_rating']['aggregate_rating']
                res_city = str(each_restaurant['restaurant']['location']['city'])
                res_lat = each_restaurant['restaurant']['location']['latitude']
                res_lon = each_restaurant['restaurant']['location']['longitude']
                cost = each_restaurant['restaurant']['average_cost_for_two']
                cuisine = each_restaurant['restaurant']['cuisines']
                zipcode = each_restaurant['restaurant']['location']['zipcode']
                print(name + " : " + str(res_rating) + " : " + str(res_city)+ " : " + str(res_lat)+ " : " + str(res_lon)+ " : " + str(cost)+ " : " + str(cuisine))
                df1 = df1.append({'restaurant_name': name,
                                  'restaurant_rating': res_rating,
                                  'resturant_city': res_city,
                                  'latitude': res_lat,
                                  'longitude': res_lon,
                                  'cost4two': cost,
                                  'cuisines':cuisine,
                                  'zipcode':zipcode
                                 }, ignore_index=True)
                listToSave.append(each_restaurant)
        counter = counter + 20
```
